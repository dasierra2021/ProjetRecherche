            Data preprocessing: renaming sujets (individuals) and surfaces


```python
# Part 1: new subjects' names

sujets_1=30
start_key = "S1"
start_value = "EXP01_001"
type="subjects"

def generate_dictionary(size,start_key,start_value,type):
    if not isinstance(size, int) or size <= 0:
        print("Please provide a positive integer parameter.")
        return None

    key_value_dict = {}

    for i in range(1, size + 1):
        if type=="subjects":
            key = f"S{i}"
            value = f"EXP01_{i:03d}"
        elif type =="surfaces":
            key = f"{i}"
            value = f"E{i:03d}"
        key_value_dict[key] = value

    return key_value_dict

#Generate the dictionary
dict_1_subjects = generate_dictionary(sujets_1,start_key,start_value,type)
```

```python
# Part 2: surface names

surfaces_1=84
start_key = "1"
start_value = "E001"
type="surfaces"

#Generate the dictionary
dict_1_surfaces = generate_dictionary(surfaces_1,start_key,start_value,type)
dict_1_surfaces = {int(key): value for key, value in dict_1_surfaces.items()}
```

```python
expe1=pd.read_excel("datasets\Expe1.xlsm",sheet_name="1")
```

```python
#Renaming values
expe1["Sujet"]=expe1.Sujet.replace(dict_1_subjects)
expe1["Surface"]=expe1.Surface.replace(dict_1_surfaces)
```

Renaming columns

```python
#Expe1
expe1_dict={
    "Sujet":"user_id1",
    "Surface":"surface_id",
    "Description":"description"
}

expe1.rename(columns=expe1_dict,inplace=True)

#Expe1_groups
expe1_groups_dict={
    "Sujet":"user_id1",
    "Groupe":"group",
    "Description":"description"
}

expe1_groups.rename(columns=expe1_groups_dict,inplace=True)

#Expe1_surfaceClass
expe1_surfaceClass_dict={
    "Sujet":"user_id1",
    "Groupe":"group",
    "Indice de Confiance":"ind_confiance",
    "Surface":"surface_id",
}

expe1_surfaceClass.rename(columns=expe1_surfaceClass_dict,inplace=True)

#Expe1_clustering
expe1_clustering_dict={
    "Surface":"surface_id",
    "Groupe":"cluster_group"
}

expe1_clustering.rename(columns=expe1_clustering_dict,inplace=True)

#Expe2
expe2_dict={
    "Numéro participant":"user_id2",
    "Suface":"codesurfE12",
    "Characteristique":"charact",
    "Valeur":"valeur",
    "Groupe":"group",
}
expe2.rename(columns=expe2_dict,inplace=True)

#surfaces_E1E2
surfaces_E1E2_dict={
    "Code":"code",
    "Hauteur":"height",
    "Diamètre (D)":"diametre",
    "Ouverture (O)":"ouverture",	
    "Espacement(SP)":"espacement"
}
surfaces_E1E2.rename(columns=surfaces_E1E2_dict,inplace=True)

#surfaces_E5
surfaces_E5_dict={
    "Code":"code",
    "Code_Expe12":"codesurf12",
    "Hauteur":"height",
    "Diamètre (D)":"diametre",
    "Ouverture (O)":"ouverture"
}

surfaces_E5.rename(columns=surfaces_E1E2_dict,inplace=True)

#expe5_charac
expe5_dict={
    "Sujet":"user_id5",
    "Surf":"codesurf5",
    "Char":"charact",
    "Val":"valeur",
    "Age":"age",
    "Main dominante":"main_domin"
}

expe5_charac.rename(columns=expe5_dict,inplace=True)
expe5_BFI.rename(columns=expe5_dict,inplace=True)
expe5_STQ.rename(columns=expe5_dict,inplace=True)
expe5_NFT.rename(columns=expe5_dict,inplace=True)
```

### Part 2: Creating a database

```python
expe1.to_sql('expe1',conn,index=False, if_exists='replace')
expe1_groups.to_sql('expe1_groups',conn,index=False, if_exists='replace')
expe1_surfaceClass.to_sql('expe1_surfaceClass',conn,index=False, if_exists='replace')
expe1_clustering.to_sql('expe1_clustering',conn,index=False, if_exists='replace')
expe2.to_sql('expe2',conn,index=False, if_exists='replace')
expe5_charac.to_sql('expe5_charac',conn,index=False, if_exists='replace')
expe5_BFI.to_sql('expe5_BFI',conn,index=False, if_exists='replace')
expe5_NFT.to_sql('expe5_NFT',conn,index=False, if_exists='replace')
expe5_STQ.to_sql('expe5_STQ',conn,index=False, if_exists='replace')
surfaces_E1E2.to_sql('surfaces_E1E2',conn,index=False, if_exists='replace')
surfaces_E5.to_sql('surfaces_E5',conn,index=False, if_exists='replace')
```
